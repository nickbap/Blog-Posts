```python
data = pd.merge(data,
                countries,
                on="Country Code",
                how="left")

data
```

```python
pd.pivot_table(data,
               index="Country",
               values=["Downloads", "Spend"],
               aggfunc=np.sum)
```

```python
dataV2 = data.fillna("Missing Country Name")

dataV2
```

```python
pd.pivot_table(data,
               index=['Country Code','Country'],
               values='Downloads',
               aggfunc=[np.sum, np.average, len])
```

```python
pivot2 = pd.pivot_table(data,
                        index="Country",
                        values=["Downloads", "Spend"],
                        aggfunc=np.sum)

pivot2["CPD"] = pivot2["Spend"] / pivot2["Downloads"]

pivot2
```
